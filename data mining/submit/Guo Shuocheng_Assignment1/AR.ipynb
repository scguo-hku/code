```python
from mlxtend import preprocessing
from mlxtend import frequent_patterns
import pandas as pd
import math
```

```python
#read order_data.csv and create a DataFrame with that content
data = pd.read_csv(r"mammographic_masses.csv",delimiter=",",header=0)
data
```

```python
#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()

#removing nan values
for i in range(0,len(d)):
    j=0
    while(True):
        if (type(d[i][j])==float and math.isnan(d[i][j])) :
            del d[i][j]
            j-=1
        j+=1
        if (j>len(d[i])-1):
            break

#adding attributes
for i in range(len(d)):
    for j in range (len(d[i])):
        d[i][j]=data.columns[j] + "=" +str(d[i][j])

            
te = preprocessing.TransactionEncoder()
te_ary = te.fit(d).transform(d)

df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
#computing frequent itemsets and association rules
frequent_itemsets = frequent_patterns.apriori(df, min_support=0.2, use_colnames=True)

a=frequent_patterns.association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

#visualizing association rules results
a[["antecedents","consequents","support","confidence"]]
```

```python
# get the total number
total_transactions = len(df)

# check if 'Age=35' and 'Severity=0' columns exist
if 'Age=35' in df.columns and 'Severity=0' in df.columns:
    # satisfy both 'Age=35' and 'Severity=0' conditions
    both_conditions = df[(df['Age=35'] == 1) & (df['Severity=0'] == 1)]
    num_both = len(both_conditions)

    # satisfy 'Age=35' condition
    age_35_conditions = df[df['Age=35'] == 1]
    num_age_35 = len(age_35_conditions)

    # compute support and confidence
    support = num_both / total_transactions
    confidence = num_both / num_age_35 if num_age_35 != 0 else 0

    # output
    print(f"support: {support:.4f}")
    print(f"confidence: {confidence:.4f}")
else:
    print("Columns not found")
```
